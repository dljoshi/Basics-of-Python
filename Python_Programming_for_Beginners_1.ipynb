# **Operators**



1.   **Aritmetic Operators**

>*   Addition:- a+b
*   Subtraction:- a-b
*   Division (Quotient):-  a/b
*   Modulus (Remainder):-  a%b
*   Exponentials:- a**b
*   Integer/floor devision:-  a//b

2.   **Assignment Operators**

>*   LHS = RHS
*   a += b --> a = a + b
*   a = 1,2,3,4
*   a,b,c,d = 1,2,3,4 

3.   **Relational/Comparision Operators**

>*   Greater Than(>):- a > b
*   Less Than(<):- a < b
*   Greater than or equal to:- a >= b
*   Less than or equal to:- a <= b
*   Equal to :- a == b
*   Not Equal to:- a != b

4.   **Logical Operators**
>*   Logical AND:- a>b and a!=b
>*   Logical OR:- a=='a' or a=='A'
>*   Logical Not:- not(a>b)
5.   **Bitwise Operators**
>*   Bitwise and (&):- a & b
>*   Bitwise or(|):- a | b
>*   Bitwise XOR(^):- a ^ b
>*   Bitwise Not(~):- ~a
>*   Bitwise Leftshift(<<) :- a << 1
>*   Bitwise Rightshift(>>):- a >> 1
6.   **Membership Operators**
*   in
>*  a=[1,2,3]; 
>*  print(3 in a) -->True
*   not in
>*  a=[1,2,3]; 
>* print(30 not in a) -->True
7.   **Identity Operators**
>* Compares the identity number not values
* is
>* a,b=1,1
>* print(a is b)
* is not
>* a,b=1,1
>* print(a is not  b)
8.   Special Operators
9.   Increment/Decrement & Condional Operators are not supported by Python

```python
# if else 
# 1. Syntax
# 2. Program to check no is even or odd..
a=int(input('Enter no'))
if a%2 == 0:
    print(a,' is even')
else:
    print(a,' is odd')
```

**Nested if-else**

```python
# Nested if else
# 1. Syntax
# 2. Program to check largest of two no...
a=int(input('Enter no1'))
b=int(input('Enter no1'))
if a>b:
    print(a,' is large')
else:
    if(a==b):
        print("Equal Nos")
    else:
        print(b,' is large')
```

**if-elif**

```python
# if elif...
# 1. Syntax
# 2. Program to check lagest of four no...
a,b,c,d=10,20,30,40
if a>b and a>c and a>d:
     print(a,' is large')
elif b>c and b>d:
     print(b,' is large')
elif c>d:
     print(c,' is large')
else:
     print(d,' is large') 
```

**while loop**

```python
# While Loop
# 1. Syntax
# 2. Program to disply no from 1 to 10

i=1
while i<=10:
    print(i)
    i+=1
print('end')
```

```python
# 3. Program to dispy no from 10 to 1.
#10 9 8 7 ....1
i=10
while i>=1:
    print(i,end=' ')
    i-=1
print('end')
```

```python
# 4. Solve the series 1+2+3+4+5+..+n
i=1
sum=0
n=int(input('Enter N'))
while i<=n:
    sum=sum+i
    i+=1
print('sum is',sum)
```

```python
# 5. Factorial of a number...
i=1
f=1
n=int(input('Enter N'))
while i<=n:
    f=f*i
    i+=1
print('Factoorial is',f)
```
