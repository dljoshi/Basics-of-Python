## Data Visualization using Matplotlib...

#### Using pyplot of matplotlib  -- provides many methods for 2D plotting of data

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

### Line Chart/graph --> plot() is used...

```python
# Line Chart/graph 
X=np.array([1,2,3,4,5])   #Data points are called markers...
Y=np.array([2,4,5,6,7])

plt.plot(X,Y)
```

##### Line Graph with more features like... Xlabel, ylabel & color.

```python
# Line Chart/graph 

X=np.array([1,2,3,4,5])
Y=np.array([2,4,5,6,7])

plt.xlabel("x-axis")   # it is also possible to give any one lable
plt.ylabel("y-axis")

plt.plot(X,Y,c="r")
```

### Sin and Cos plot using plot()

```python

# Sin and Cos plot using plot()
x = np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
plt.plot(x,a,c='g',linewidth=5)  #cos
plt.plot(x,b,c='r',linewidth=2, linestyle='dotted')  #sin
```

#### using linewidth & linestyle
#### Color Code --> b -bule, r-red, g-green, y-yello, k-black, c-cyan, m-magenta
#### linestyle --> solid, dashed, dotted, dashdot

```python
# Using Line Width... 
x = np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
plt.plot(x,a,c='g',linewidth=2)  #cos
plt.plot(x,b,c='r',linewidth=2, linestyle='dashdot')  #sin
```

#### Changing Marker Type --- 
#### marker = it can be --> . , o + x D d s p * v ^ > < etc
#### we can use markersize, markeredgecolor.

```python
# Line Chart/graph 
X=np.array([1,2,3,4,5])
Y=np.array([2,4,5,6,7])

plt.xlabel("x-axis")   # it is also possible to give any one lable
plt.ylabel("y-axis")

plt.plot(X,Y,c="r",marker='D',markersize=10,markeredgecolor='green')
```

```python
%matplotlib inline
# magic function...
X=np.array([1,2,3,4,5])
Y=np.array([2,4,5,6,7])

plt.xlabel("x-axis")   
plt.ylabel("y-axis")

plt.plot(X,Y,c="k",marker='>',markersize=10,markeredgecolor='cyan')
```

#### Creating multiple bar charts...

```python
X=np.array([1,2,3,4,5])

Y1=np.array([2,4,5,6,7])
Y2=np.array([3,6,5,6.7,7.1])

plt.xlabel("x-axis")   
plt.ylabel("y-axis")

plt.bar(X,Y1,color='k', width=0.3)
plt.bar(X+0.3,Y2,color='b',width=0.3)
```
